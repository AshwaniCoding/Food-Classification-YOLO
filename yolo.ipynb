```python
import tensorflow as tf
from yolov8m.pt import create_yolo_model  # Import function to create YOLO model
from load_yolo_weights import load_yolo_weights  # Import function to load YOLO weights

# Create an instance of the YOLO model in TensorFlow/Keras
yolo_model = create_yolo_model()

# Load YOLO weights into the TensorFlow/Keras model
yolo_weights_path = "/content/drive/MyDrive/Food Classification YOLO/runs/detect/train/weights"
load_yolo_weights(yolo_model, yolo_weights_path)

# Save the model as .h5 file
tf.keras.models.save_model(yolo_model, 'yolo_model.h5')
```
